```python
import statistics
import random

kitaplar = [ 
{"isim": "Veri Bilimi 101"       , "yazar": "Ali"  , "tur": "Bilim"   , "satis": 1200, "yil": 2021}, 
{"isim": "Python ile Yapay Zeka" , "yazar": "Ayşe" , "tur": "Bilim"   , "satis": 950 , "yil": 2020}, 
{"isim": "İstatistik Temelleri"  , "yazar": "Ali"  , "tur": "Akademik", "satis": 700 , "yil": 2019}, 
{"isim": "Makine Öğrenmesi"      , "yazar": "Can"  , "tur": "Bilim"   , "satis": 1800, "yil": 2022}, 
{"isim": "Veri Görselleştirme"   , "yazar": "Deniz", "tur": "Sanat"   , "satis": 400 , "yil": 2018}, 
{"isim": "Matematiksel Modelleme", "yazar": "Ali"  , "tur": "Akademik", "satis": 1500, "yil": 2021}, 
{"isim": "Bilgi Toplumu"         , "yazar": "Ayşe" , "tur": "Sosyal"  , "satis": 600 , "yil": 2022} 
]

def en_cok_satan(kitaplar):
    cok_satan = kitaplar[0]

    for kitap in kitaplar:
        if cok_satan["satis"] < kitap["satis"]:
            cok_satan = kitap
    
    return cok_satan["isim"]

def yazar_satislari(kitaplar):
    tekil_yazar = list(dict.fromkeys([kitap["yazar"] for kitap in kitaplar]))
    yazar_dict = {}

    for yazar in tekil_yazar:
        yazar_dict[yazar] = sum([kitap["satis"] for kitap in kitaplar if kitap["yazar"] == yazar])
    return yazar_dict

def listele(secim , kitaplar):    
    yazarlar = list(dict.fromkeys([kitap[secim] for kitap in kitaplar]))
    return yazarlar

def satis_listele(adet):
    cok_satanlar = [kitap["isim"] for kitap in kitaplar if kitap["satis"] > adet]
    return cok_satanlar

def yil_filtre(yil):
    sonrasi_kitaplar = list(filter(lambda k: k["yil"] > yil, kitaplar))
    return sonrasi_kitaplar

def artis_uygula(oran):
    satis_artis = list(map(lambda k: k["satis"] * oran, kitaplar))
    return satis_artis

def sirala(kitaplar):
    sirali = sorted(kitaplar, key=lambda k: k["satis"], reverse=True)
    for k in sirali:
        print(k["isim"], "-", k["satis"])

def ort_satis(kitaplar):
    toplam = 0

    for kitap in kitaplar:
        toplam += kitap["satis"]
    
    return(toplam/len(kitaplar))
    
def cok_satan_tur(kitaplar):
    tekil_tur = list(dict.fromkeys([kitap["tur"] for kitap in kitaplar]))
    yazar_dict = {}

    for yazar in tekil_tur:
        yazar_dict[yazar] = sum([kitap["satis"] for kitap in kitaplar if kitap["tur"] == yazar])
    return yazar_dict

def sapma(kitaplar):
    satis = []

    for kitap in kitaplar:
        satis.append(kitap["satis"])
    
    return statistics.stdev(satis)


egitim_boyut = int(len(kitaplar) * 0.7)
egitim = random.sample(kitaplar, egitim_boyut)
test = [k for k in kitaplar if k not in egitim]

def yazar_satis_ort(egitim):
    yazar_satis = yazar_satislari(egitim)
    toplam = 0

    for yazar in yazar_satis.values():
        toplam += yazar
    return (toplam/len(yazar_satis))

def ort_ustu_satis(egitim):
    ort = yazar_satis_ort(egitim)

    for kitap in egitim:
        if kitap["satis"] > ort:
            print([kitap])



print("En çok satan kitap:           " , en_cok_satan(kitaplar))
print("Yazar satışları:              ", yazar_satislari(kitaplar))
print("Türler:                       " , listele("tur" , kitaplar))
print("1000'den fazla satan kitaplar:" , satis_listele(1000))
print("Ortalama satış:               " , ort_satis(kitaplar))
print("Standart sapma:               " , sapma(kitaplar))

print("*********** Eğitim Test Ayrımı ***********")
print("************** Eğitim %70  ***************")
print("Ortalama:       " , statistics.mean(listele("satis", egitim)))
print("Mod:            " , statistics.mode(listele("satis", egitim)))
print("Medyan:         " , statistics.median(listele("satis", egitim)))
print("Standart Sapma: " , statistics.stdev(listele("satis", egitim)))
print("*************** Test %30  ****************")
print("Ortalama:       " , statistics.mean(listele("satis", test)))
print("Mod:            " , statistics.mode(listele("satis", test)))
print("Medyan:         " , statistics.median(listele("satis", test)))
print("Standart Sapma: " , statistics.stdev(listele("satis", test)))
```
