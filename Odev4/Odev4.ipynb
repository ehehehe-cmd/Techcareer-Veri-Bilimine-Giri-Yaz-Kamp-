```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
matris = np.random.randint(0, 101, size=(5, 5))


print(matris)
print("Standart sapma: " , matris.std())
print("Min değer: " , matris.min() , " Max değer: " , matris.max())
print("Köşegen toplamı: " , matris.diagonal().sum())
```

```python
data = {
    "Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Yaş": [20, 21, 19, 22, 20],
    "Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
    "Matematik": [70, 60, 80, 90, 55],
    "Fizik": [65, 75, 70, 85, 60],
    "Kimya": [80, 85, 65, 95, 70]
}

df = pd.DataFrame(data)

df.to_csv("ogrenci_notlari.csv", index=False, encoding="utf-8")

print(df)
```

```python
print("Matematik ortalaması: " , df["Matematik"].mean())
print("Fizik ortalaması: " , df["Fizik"].mean())
print("Kimya ortalaması: " , df["Kimya"].mean())

print("En yüksek matematik notu alan kişi:" , df.loc[df["Matematik"].idxmax(), "Öğrenci"])

df["Ortalama"] = df[["Matematik", "Fizik", "Kimya"]].mean(axis=1)

print(df)
```

```python
print(df.groupby("Bölüm")[["Ortalama"]].mean())

print(df[df["Ortalama"] > 70])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

dersler = ["Matematik", "Fizik", "Kimya"]

for i, ders in enumerate(dersler):
    sns.histplot(data=df, x=ders, bins=10, kde=True, ax=axes[i])
    axes[i].set_title(f"{ders} Not Dağılımı")

plt.tight_layout()
plt.show()
```

```python
bolum_ortalama = df.groupby("Bölüm")[["Matematik", "Fizik", "Kimya", "Ortalama"]].mean().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=bolum_ortalama, x="Bölüm", y="Ortalama", palette="viridis")
plt.title("Bölümlere Göre Ortalama Başarı")
plt.show()
```
